```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## 1 - Definindo os parâmetros do modelo

```python
n_simulations = 1000
processing_time_mean = 5
processing_time_std = 2
interarrival_time_mean = 10
interarrival_time_std = 5
```

## 2 - Simulando o tempo de processamento e tempo entre chegadas para cada posição

```python
processing_time = np.random.normal(processing_time_mean, processing_time_std, n_simulations)
interarrival_time = np.random.normal(interarrival_time_mean, interarrival_time_std, n_simulations)
```

## Calculando o tempo de espera na fila para cada simulação

```python
queue_time = np.zeros(n_simulations)
for i in range(1, n_simulations):
    queue_time[i] = max(queue_time[i-1] + processing_time[i-1] - interarrival_time[i], 0)
```

## Calculando a média e o desvio padrão do tempo de espera na fila



## Plot do Gráfico

```python
ax = pd.DataFrame({'Queue Time': queue_time}).plot()
ax.set_xlabel('Position')
ax.set_ylabel('Queue Time (seconds)')
plt.show()
```

```python
mean_queue_time = np.mean(queue_time)
std_queue_time = np.std(queue_time)

print("Média do tempo de espera na fila:", mean_queue_time)
print("Desvio padrão do tempo de espera na fila:", std_queue_time)
```
